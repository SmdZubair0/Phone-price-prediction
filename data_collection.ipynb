```python
# important features
features = [ 'title',
 'rating',
 'image',
 'price',
 'In The Box',
 'Model Name',
 'Color',
 'SIM Type',
 'Touchscreen',
 'Quick Charging',
 'Display Size',
 'Resolution',
 'GPU',
 'Operating System',
 'Processor Brand',
 'Processor Type',
 'Processor Core',
 'Primary Clock Speed',
 'Internal Storage',
 'RAM',
 'Expandable Storage',
 'Primary Camera',
 'Secondary Camera',
 'HD Recording',
 'Video Recording Resolution',
 'Network Type',
 'Battery Capacity',
 'Battery Type',
 'Warranty Summary',
 'JAVA Support']
```

```python
SPECS = {}

for i in features:
    SPECS[i] = []
```

```python
SPECS
```

```python
# for i,link in enumerate(phone_links[:5]):
phone_links = np.load('files/phone_links.npy')

i = 0
while i < len(phone_links):
    try:
        link = phone_links[i]
        response = requests.get(link)
        soup  = BeautifulSoup(response.text,'html.parser')

        for k in SPECS.keys():
            SPECS[k].append(np.nan)
            
        rows = soup.find_all(class_='WJdYP6 row')
        
        title = soup.find('span',class_='VU-ZEz').text
        rating = soup.find(class_='XQDdHH').text
        image = soup.find(class_='DByuf4 IZexXJ jLEJ7H').get('src')
        price = soup.find(class_='yRaY8j A6+E6v').text

        SPECS['title'][-1] = title
        SPECS['rating'][-1] = rating
        SPECS['image'][-1] = image
        SPECS['price'][-1] = price

        for row in rows:
            att = row.find(class_='+fFi1w col col-3-12').text
            val = row.find(class_='HPETK2').text

            if att in SPECS.keys():
                SPECS[att][-1] = val
        
        time.sleep(7)

        print(i)

        i += 1
    except Exception as e:
        print(e)
        for k in SPECS.keys():
            SPECS[k].pop()
        time.sleep(7)
```
